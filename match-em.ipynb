```python
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import numpy as np
import base64
from numpy.random import default_rng
import random


rng= np.random.default_rng
#defines the random number generator utilized to index values 

CusTrans = pd.read_csv("../input/h-and-m-personalized-fashion-recommendations/transactions_train.csv")
CusAge = pd.read_csv('../input/ages-20-25-30-35/Ages x5.csv')
article= pd.read_csv("../input/articles/clean Article.csv")

ra = pd.merge(CusTrans,CusAge, on=['customer_id'], how='inner')
res = pd.merge(ra,article, on=['article_id'], how='inner')

length_of_dataset= len(res)

list_index_vals=[]

x_count=0

for i in range(len(res)):
    list_index_vals.append(x_count)
    x_count=x_count+1
    
print(length_of_dataset)


res.insert(loc=1,
          column= "Random Index Number",
          value=random.sample(list_index_vals,len(res))
           
          )
#     random.sample())
print(res)

res.to_csv("new.csv", index=False)
```
